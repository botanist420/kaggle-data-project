```python
# Curry regular season stats
file_path = "../dataset/kaggle_curry/Stephen Curry Stats.csv"
```

```python
df_curry = pd.read_csv(file_path)
df_curry.head()
```

```python
# extract Season_year=2020-2021
# df_curry.Season_year.unique()

df_2021 = df_curry[df_curry.Season_year == '2020-2021']

df_2021.FT.head()

df_2021[['free_throw_made', 'free_throw_attempt']] = df_2021.FT.str.split(pat='-', expand=True).astype(int)

ftm_total, fta_total = df_2021.free_throw_made.sum(), df_2021.free_throw_attempt.sum()
ft_rate = ftm_total / fta_total

print(ftm_total, fta_total)
print(ft_rate.round(5))
```
